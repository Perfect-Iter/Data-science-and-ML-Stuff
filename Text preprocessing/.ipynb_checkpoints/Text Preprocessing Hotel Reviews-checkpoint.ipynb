```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
df.at[2, 'Review']
```

```python
nltk.download('stopwords')
```

```python
def clean_message(message, stemmer = PorterStemmer(), 
                  stop_words = set(stopwords.words('english'))):
    
    
    words = word_tokenize(message)
    
    filtered_words = []
    
    for word in words:
        
        if word not in stop_words and word.isalpha():
            filtered_words.append(stemmer.stem(word))
    
    return filtered_words
```

```python
nested_list = df['Review'].apply(clean_message)
```

```python
nested_list
```

```python
df.dtypes
```

```python
df['Review'] = df['Review'].apply(clean_message)
```

```python
df.head()
```

```python
flat_list = [item for sublist in nested_list for item in sublist]
```

```python
flat_list
```

```python
unique_words = pd.Series(flat_list).value_counts()
print('No. of Unique Words', unique_words.shape[0])
unique_words.head(20)
```

```python
word_cloud = WordCloud( background_color='white', colormap='winter')
word_cloud.generate(','.join(flat_list))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
